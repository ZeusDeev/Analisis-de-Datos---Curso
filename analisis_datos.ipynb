### Series y DataFrame con pandas.

Una serie es una estructura de datos unidimencional que permite almacenar cualquier tipo de datos.

Por otro lado los dataframe son estructura de datos bidimencionales que pueden contener multiples columnas y filas de datos y ambos pueden ser creados apartir de listas,diccionarios o arreglos de numpy.

```python
import pandas as pd
```

```python
#Series
numeros = [1,2,3,4]
series = pd.Series(numeros)
print(series)
```

 ## DataFrame con Pandas:


 En un **dataframe** podemos crear desde un diccionario, colocando primeramente el indice de la tabla, es decir el encabezado de nuestra columna, seguido de una lista para colocar nuestros valores.

```python
#DataFrame

datos = {
    'Nombre':['Juan','Maria','Pedro','Abraham'],
    'Edad':[15,20,12,4],
    'Ciudad':['Lima','Arequipa','Lima','Carrillo Puerto']
}


#Genera un DataFrame apartir de un diccionario.

df = pd.DataFrame(datos)
print(df)
```

###Exporta Dataframe como archivo csv.

```python
df.to_csv('datos.csv')
```

### Importa un DataFrame
